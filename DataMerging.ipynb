```python
import pandas as pd

genre_studios = pd.read_csv('Data/movie_gross_df_CLEAN.csv')
imdb_titles = pd.read_csv('Data/imdb_title_basics.csv')
imdb_ratings = pd.read_csv('Data/imdb_title_ratings_df_CLEAN.csv')
money = pd.read_csv('Data/FinalCleaning.csv')
```

```python
imdb_titles.head()
```

```python
imdb_titles['Link'] = imdb_titles['original_title'].str.replace(r'[^\w\s]+', '')
imdb_titles
```

```python
imdb_titles['Link'] = imdb_titles['Link'].str.lower()
```

```python
imdb_titles['start_year_str'] = imdb_titles['start_year'].astype(str)
```

```python
imdb_titles['Link'] = imdb_titles['Link'] + imdb_titles['start_year_str']
```

```python
imdb_titles
```

```python
imdb_titles_w_studio = imdb_titles.merge(genre_studios, on = "Link")
```

```python
imdb_titles_w_studio
```

```python
imdb_titles_w_studio.to_csv('imdb_titles_w_studio.csv')
```

```python
imdb_title_rating_studio = imdb_titles_w_studio.merge(imdb_ratings, on = 'tconst' )
```

```python
imdb_title_rating_studio
```

```python
money
```

```python
full_imdb_data = imdb_title_rating_studio.merge(money, left_on='Link', right_on='nameYear')
```

```python
full_imdb_data
```

```python
full_imdb_data.to_csv('Complete_CLEAN_IMDB_Data.csv')
```
